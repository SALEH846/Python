```python
import os
```

```python
# Get current working Directory
print(os.getcwd())
```

```python
# Methods which you can use from OS module.
print(dir(os))
```

```python
# You can change current working directory.
os.chdir('/home/saleh/Documents/')
# Now check your working directory.
print(os.getcwd())
```

```python
# You can list the directories(folders) and files in your current working directory.
print(os.listdir())
print("\n") # To create a blank line between the two.
# You can also pass path to specify the directory of which you want to see the list files and directories.
print(os.listdir('/home/saleh/Documents/'))
```

```python
# For making directories we have two methods, we will see both of them.
# The first one is mkdir. You can use it when to you want to create a single directory. If you pass just the name 
# of directory it will create it in your current working directory by default but if you want to specify the path 
# you can also do it.
os.mkdir('os_module_1')
os.mkdir('/home/saleh/Documents/Python_jupyter/os_module_2')
# The second method for this purpose is makedirs. Both of them are pretty similar but makedirs is used when you 
# want to create a directory which is deep inside of some directories and they all not exist so it will create 
# all of them.
os.makedirs('os_module__3/sub_directory_1/sub_directory_2')
os.makedirs("/home/saleh/Documents/Python_jupyter/os_module_4/sub_dir_1/sub_dir_2")
```

```python
# For deleting directories we have two methods, we will see both of them.
# The first one is rmdir. You can use it when to you want to remove a single directory. If you pass just the name 
# of directory it will remove it in your current working directory by default, if exists but if you want to 
# specify the path you can also do it.
os.rmdir('os_module_1/')
os.rmdir("/home/saleh/Documents/Python_jupyter/os_module_2")
# The second method for this purpose is removedirs. Both of them are pretty similar but removedirs is used when  
# you want to remove a directory which is deep inside of some directories and this method deletes all the 
# intermediate directories between your current working directory and the directory you want to delete.
os.removedirs('os_module__3/sub_directory_1/sub_directory_2')
```

```python
# You can also rename your files by using OS module. For example, I have a text file named test.txt in my current
# working directory and I wannt to change its name to demo.txt. Let's see how we can do it.
os.rename('test.txt', 'demo.txt')
```

```python
# If you want to print some information about file you can use stat method.
print(os.stat('demo.txt'))
# The result looks gebbrish but it has some useful things in it.
# For example: st_size attribute is telling us that the size of file is 0 bytes. We can print alone in this way.
print(os.stat('demo.txt').st_size)
# The other useful thing is st_mtime attribute which is telling us the modification time and date but it is not 
# in a human readable format but we can convert it in human readable format by using python's datetime module.
# The format which it is already in is called time stamp. Let's see how.
mod_time = os.stat('demo.txt').st_mtime
from datetime import datetime
print(datetime.fromtimestamp(mod_time))
```

```python
# If you want to see all the directories and files in a specific directory and also all the directories and files 
# inside each of it you can use walk method.
for dirpath, dirnames, filenames in os.walk('/home/saleh/Documents/'):
    print('Current path:', dirpath)
    print('Directories:', dirnames)
    print('Files', filenames)
    print()
```

```python
# If you want to see your home directory, you can use os.environ method.
print(os.environ.get("HOME"))
```

```python
# Let's I want to create a new file called test.txt in my home directory by conacatenating file paths such as 
# os.environ.get("HOME") + test.txt. This will not work because the home directory has not a forward slash at the
# end of it so we end up in getting something like this.
file_path = os.environ.get("HOME") + 'test.txt'
print(file_path)
# So to prevent this we can use the join method of os.path module.
file_path = os.path.join(os.environ.get("HOME"), 'test.txt')
print(file_path)
```

```python
# If you want to check that wether a certain exists or not you can use exists method of os.path module.
print(os.path.exists('/tmp/test.txt'))
```

```python
# If you want to check wether it is file or a directory.
print(os.path.isfile('/home/saleh/Documents/freelance details'))
print(os.path.isdir('/home/saleh/Documents/freelance details'))

print(os.path.isdir('/home/saleh/Documents/python_jupyter/'))
print(os.path.isfile('/home/saleh/Documents/python_jupyter/'))
```

```python
# Another useful function of this module is splitext. It split the file's root name and its extension.
print(os.path.splitext('/tmp/test.txt'))
```

```python
# All the functions you can try from os.path module.
print(dir(os.path))
```

```python
# If you want to fetch the base name from thee entire path.
print(os.path.basename('/home/saleh/Documents/freelance details.txt'))
# If you want to fetch the root name from the entire path.
print(os.path.dirname('/home/saleh/Documents/freelance details.txt'))
# If you want both in the form of tuple you can use split  method.
print(os.path.split("/home/saleh/Documents/freelance details.txt"))
```
